## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

Entities in a table can relate to other entities in 1-1, 1-many and many-to-many relationships.

In 1-1 relationships, an entity only references one other entity in another table. 

In 1-many relationships, an an entity can be connected to many different objects from another table through a foreign key on the secondary (many) table. But entities int hat table can only reference one object.

In many-many relationships, any entity in either table can reference any number of objects in another table, and usually require an associative table to map the many references.
